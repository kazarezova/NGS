4. Оцените экспрессию генов в каждом образце при помощи
Rsubread

```r
if (!require("BiocManager", quietly = TRUE))
    install.packages("BiocManager")

BiocManager::install("Rsubread")
```

```r
library(Rsubread)
```

```r
bams <- c('B14.5','B15.5','B17.5','B20','B34','C14.5','C15.5','C17.5','C20','C34')
for (bam in bams) {
  bamfile <- paste(bam,'.stringtie.s.bam', sep = "", collapse = "")
  tsvfile <- paste(bam,'.tsv', sep = "", collapse = "")
  fc <- featureCounts(bamfile,annot.ext = 'merged.gtf',isGTFAnnotationFile=TRUE)
  write.table(
    x=data.frame(fc$annotation[,c("GeneID","Length")],
      fc$counts,
      stringsAsFactors=FALSE),
    file=tsvfile,
    quote=FALSE,
    sep="\t",
    row.names=FALSE)
}
```
